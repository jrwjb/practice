```python
import re
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('P20190100135_1/peptides.txt', sep='\t')
df2 = pd.read_csv('P20190100135_2/peptides.txt', sep='\t')
df3 = pd.read_csv('P20190100135_3/peptides.txt', sep='\t')
df4 = pd.read_csv('P20190100135_4/peptides.txt', sep='\t')
df5 = pd.read_csv('P20190100135_5/peptides.txt', sep='\t')
# df1.tail()
```

```python
df5.head()
```

```python
def clean(df, index):
    # drop_list
    columns = df.columns.tolist()
    intensity_list = [i for i in columns[columns.index('Intensity'):] if re.search('Intensity|iBAQ', i, re.I)]
    newcolumns = intensity_list
    newcolumns.insert(0, columns[0])
    if index != 1:
        df = df[newcolumns]
    else:
        df = df[columns[:columns.index(intensity_list[-1]) + 1]]
    return df

df1 = clean(df1, 1)
df2 = clean(df2, 2)
df3 = clean(df3, 3)
df4 = clean(df4, 4)
df5 = clean(df5, 5)
```

```python
df3.head()
```

```python
df_merge1 = pd.merge(df1, df2, how='outer', on='Sequence', suffixes=('_1', '_2'))
df_merge2 = pd.merge(df_merge1, df3, how='outer', on='Sequence')
df_merge3 = pd.merge(df_merge2, df4, how='outer', on='Sequence', suffixes=('_3', '_4'))
df_merge4 = pd.merge(df_merge3, df5, how='outer', on='Sequence')
df_merge4.to_csv('peptide_merge.csv', index=False)
```
